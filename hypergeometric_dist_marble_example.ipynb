```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# settings
sns.set_style('darkgrid')
```

```python
marble_bag = 10 # N
red_marbles = 3 # k
marbles_drawn = 3 # n
```

```python
rv = scipy.stats.hypergeom(marble_bag, red_marbles, marbles_drawn)
```

```python
x = np.arange(0, marbles_drawn+1)
```

```python
pmf_marbles = rv.pmf(x)
```

```python
pmf_marbles = pmf_marbles * 100
pmf_marbles
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, pmf_marbles, 'bo')
ax.vlines(x, 0, pmf_marbles, lw=2)
ax.set_xlabel('# of marbles drawn')
ax.set_ylabel('hypergeom PMF')
ax.set_xticks(x)
plt.show()
```
